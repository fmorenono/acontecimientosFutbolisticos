```python
#Carga de datos estadisticas


import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns', None)
partidos_2015_df = pd.read_csv("2015_2016_spain_estadisticas_v3.csv", sep = ';')

partidos_2016_df = pd.read_csv("2016_2017_spain_estadisticas_v3.csv", sep = ';')

partidos_2018_df = pd.read_csv("2018_2019_spain_estadisticas_v3.csv", sep = ';')

partidos_2017_df = pd.read_csv("2017_2018_spain_estadisticas_v3.csv", sep = ';')

partidos_2015_df['temporada']='2015'
partidos_2016_df['temporada']='2016'
partidos_2017_df['temporada']='2017'
partidos_2018_df['temporada']='2018'
partidos_df=pd.concat([partidos_2015_df,partidos_2016_df, partidos_2017_df, partidos_2018_df])
```

```python
#Carga de datos metereologicos AeMET

partidos_condiciones_aemet_2015_df = pd.read_csv("weatherAEMET_2015.csv", sep = ';')
partidos_condiciones_aemet_2016_df = pd.read_csv("weatherAEMET_2016.csv", sep = ';')
partidos_condiciones_aemet_2017_df = pd.read_csv("weatherAEMET_2017.csv", sep = ';')
partidos_condiciones_aemet_2018_df = pd.read_csv("weatherAEMET_2018.csv", sep = ';')
partidos_condiciones_aemet_df=pd.concat([partidos_condiciones_aemet_2015_df, partidos_condiciones_aemet_2016_df, partidos_condiciones_aemet_2017_df,\
                                         partidos_condiciones_aemet_2018_df])

correlated_features = set()
correlation_matrix = partidos_condiciones_aemet_2018_df.corr()

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.7:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)
correlated_features
#Seleccion de los features significativos
partidos_condiciones_aemet_df
partidos_condiciones_aemet_df=partidos_condiciones_aemet_df[['ID','aemet_tmax','aemet_tmin','aemet_sol','aemet_prec','aemet_racha','aemet_tmax_vis','aemet_tmin_vis','aemet_sol_vis',\
                                                             'aemet_prec_vis','aemet_racha_vis']]
```

```python
partidos_condiciones_aemet_df.isnull().sum(axis = 0)
```

```python



partidos_condiciones_aemet_df['aemet_tmax'].fillna(value=partidos_condiciones_aemet_df['aemet_tmax'].mean(), inplace=True)
partidos_condiciones_aemet_df['aemet_tmin'].fillna(value=partidos_condiciones_aemet_df['aemet_tmin'].mean(), inplace=True)
partidos_condiciones_aemet_df=partidos_condiciones_aemet_df.drop(['aemet_sol'], axis = 1)
partidos_condiciones_aemet_df['aemet_prec'].fillna(value=partidos_condiciones_aemet_df['aemet_prec'].mean(), inplace=True)
partidos_condiciones_aemet_df['aemet_prec_vis'].fillna(value=partidos_condiciones_aemet_df['aemet_prec_vis'].mean(), inplace=True)
partidos_condiciones_aemet_df['aemet_racha'].fillna(value=partidos_condiciones_aemet_df['aemet_racha'].mean(), inplace=True)
partidos_condiciones_aemet_df=partidos_condiciones_aemet_df.drop(['aemet_sol_vis'], axis = 1)
partidos_condiciones_aemet_df['aemet_racha_vis'].fillna(value=partidos_condiciones_aemet_df['aemet_racha_vis'].mean(), inplace=True)
#partidos_condiciones_aemet_df.describe()
partidos_condiciones_aemet_df.isnull().sum(axis = 0)
```

```python
#Diferencia de temperaturas medias
partidos_condiciones_aemet_df['aemet_t_dif']=(((partidos_condiciones_aemet_df['aemet_tmax']+partidos_condiciones_aemet_df['aemet_tmin'])/2)-\
((partidos_condiciones_aemet_df['aemet_tmax_vis']+partidos_condiciones_aemet_df['aemet_tmin_vis'])/2))
#Diferencia de rango de temperaturas
partidos_condiciones_aemet_df['aemet_t_dif_rango']=((partidos_condiciones_aemet_df['aemet_tmax']-partidos_condiciones_aemet_df['aemet_tmin'])-
(partidos_condiciones_aemet_df['aemet_tmax_vis']-partidos_condiciones_aemet_df['aemet_tmin_vis']))
#Rango temperaturas home
partidos_condiciones_aemet_df['aemet_rango']=(partidos_condiciones_aemet_df['aemet_tmax']-partidos_condiciones_aemet_df['aemet_tmin'])
#Rango temperaturas away
partidos_condiciones_aemet_df['aemet_rango_vis']=(partidos_condiciones_aemet_df['aemet_tmax_vis']-partidos_condiciones_aemet_df['aemet_tmin_vis'])
#Diferencia de vel maxima (rachas)

partidos_condiciones_aemet_df['aemet_v_dif']=(partidos_condiciones_aemet_df['aemet_racha']-partidos_condiciones_aemet_df['aemet_racha_vis'])

partidos_condiciones_aemet_df
```
